```python
data = data.drop_duplicates()
```

```python
data["product_url"].nunique()
```

```python
import ast

data["test_type"] = data["test_type"].apply(ast.literal_eval)
data["Job_levels"] = data["Job_levels"].apply(ast.literal_eval)
```

```python
data.head()
```

```python
test_type_map = {
    "A": "Ability & Aptitude",
    "B": "Biodata & Situational Judgement",
    "C": "Competencies",
    "D": "Development & 360",
    "E": "Assessment Exercises",
    "K": "Knowledge & Skills",
    "P": "Personality & Behavior",
    "S": "Simulations"
}
data["test_type"] = data["test_type"].apply(lambda types: [test_type_map.get(t, t) for t in types])
data.head()
```

```python
data = data.rename(columns={"Remote_testing_available": "remote_testing_available"})
data.head()
```

```python
data["remote_testing_available"] = data["remote_testing_available"].map({True: "Yes", False: "No"})
```

```python
data.head()
```

```python
data["product_description"] = (data["product_description"].str.replace(r"\r\n|\n|\t", " ", regex=True))

data["product_description"] = (data["product_description"].str.replace(r"\s+", " ", regex=True).str.strip())

data.head()
```

```python
data.to_csv("products_cleaned.csv", index=False)
data.to_json("products_cleaned.json", orient="records", indent=2)
```
